```python
from matplotlib import pyplot as plt
```

```python
x = [1,2,3,41]
y = [23,32,43,64]
plt.plot(x,y)
```

```python
subject = ['Maths','Science','Social','English','Computer Science']
student1 = [90,89,70,98,78]
student2 = [97,56,89,92,90]
plt.bar(subject,student1,color = 'g',label = "Nisha")
plt.bar(subject,student2,color = 'k', label = "Rahul")
plt.legend()
plt.title("Nisha vs Rahul")
plt.xlabel("Subjects")
plt.ylabel('Marks')
```

```python
subject = ['Maths','Science','Social','English','Computer Science']
values = [90,89,70,98,78]
plt.pie(values,labels = subject)
```

```python
subject = ['Maths','Science','Social','English','Computer Science']
values = [90,89,70,98,78]
x_values = [0.1,0,0,0.2,0]
plt.pie(values,labels = subject,explode = x_values)
```
